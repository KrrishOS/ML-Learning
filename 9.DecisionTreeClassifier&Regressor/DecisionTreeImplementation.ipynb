```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris['target']
```

```python
X = pd.DataFrame(iris['data'],columns = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'])
y = pd.DataFrame(iris['target'])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2 ,random_state=555)
```

```python
## no need for standraization in DTC
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treeclassifier,filled = True)
```

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
cm = confusion_matrix(y_pred,y_test)
acc = accuracy_score(y_pred,y_test)
cr = classification_report(y_pred,y_test)
print(cm,acc,cr)
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
df = pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
df.head()
```

```python
df.feature_names
```

```python
df
```

```python
X = df
y = dataset.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2 ,random_state=555)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
dregressor = DecisionTreeRegressor()
dregressor.fit(X_train,y_train)
```

```python
y_pred5 = dregressor.predict(X_test)
```

```python
regressor = DecisionTreeRegressor()
```

```python
param = {
    'criterion': ['squared_error', 'friendman_mse', 'absoulute_error'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5,10,15,20,25],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
## using Hyperparameter tuning for better results
grid1 = GridSearchCV(regressor,param_grid= param,cv=5,scoring='neg_mean_squared_error')
```

```python
grid1.fit(X_train,y_train)
```

```python
grid1.best_params_
```

```python
y_pred23 = grid1.predict(X_test)
```

```python

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2= r2_score(y_test,y_pred23)
mae = mean_absolute_error(y_test,y_pred23)
mse= mean_squared_error(y_test,y_pred23,)
print(r2,mae,mse)
```
